```python
for n in range(9999, 999, -1):
    b = str(n)
    l = int(b[0]) + int(b[1]) 
    r = int(b[2]) + int(b[3])
    s = ''
    if l > r:
        s = str(r) + str(l)
    else:
        s = str(l) + str(r)
    if s == '117':
        print(n)
        break
```

```python
from itertools import *
a = list(product('', repeat=5))
c = 0
for i in range(0, len(a), 1):
    if a[i].count('') == 1:
        c += 1
print(c) 
```

```python
def f(n):
    if n == 0:
        return 0
    if n > 0 and n % 2 == 0:
        return f(n//2) 
    if n % 2 == 1:
        return 1 + f(n - 1)
c = 0
for n in range(1, 501, 1):
    if f(n) == 8:
        c += 1
print(c)
```

```python
def f(x, y):
    if x > y or x == 17:
        return 0
    if x == y:
        return 1
    else:
        return f(x + 1, y) + f(x * 2, y)
print(f(1,10)*f(10,21))
```
