```python
a = open('24.txt').readline()
#a = 'xzctyvvbasdfghj'
f = []
c = 0
b = []
for i in range(0, len(a), 1):
    if a[i] in f:
        f = []
        b.append(c)
        c = 1
    else:
        f.append(a[i])
        c += 1
b.append(c)        #мы записываем последний счетчик (с) в массив b
print(max(b))
        
```

```python
a = open('9.txt').readlines()
ch = 0
for i in range(0, len(a), 1):
    u1 = 0
    u2 = 0
    u3 = 0
    b = list(map(int,a[i].split()))
    c = 0
    n = []
    h = []
    for el in b:
        if b.count(el) == 1:
            c += 1
    if c == 5:                   
        u1 = 1

    for el in b:
        if el % 2 == 0:
            h.append(el)
        else:
            n.append(el)
    if len(n) > len(h):
        u2 = 1
    if sum(n) < sum(h):
        u3 = 1
    if u3 == 1 and u2 == 1 and u1 == 1:
        ch += 1 
print(ch)
 
```
