```python
f = open('24 (1).txt').readline()
alp = sorted('QWERTYUIOPASDFGHJKLZXCVBNM')[:13] #13  
c1 = []

for el in alp:
    a = f.split(el) # '1 2 3' -> f.split() -> ['1', '2', '3']       '1A2A3' -> f.split('A') -> ['1', '2', '3']
    for el1 in a:
        c1.append(len(el1))
    
print(max(c1)+2)
```

```python
f = open('24_2.txt').readline()
c = []
c1 = 0

for i in range(0, len(f) - 1, 1):
    if f[i] not in 'ABC' or f[i+1] not in 'ABC':
        c1 += 1
    else:
        c.append(c1)
        c1 = 1 #подсчет с 1 потому что в начале подсчет пары 'a{az}xdvf'
print(max(c))

```

```python
from itertools import *
a = list(product('АБВГД', repeat=5))
c = 0
for i in range(0, len(a), 1):
    s = ''.join(a[i])
    if a[i][0] != 'А' and ('АА'not in s or 'ББ'not in s or 'ВВ'not in s or 'ГГ'not in s or 'ДД'not in s):
        c += 1
print(c)
        
```
