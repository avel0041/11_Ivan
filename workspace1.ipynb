```python
d = []
for n in range(10, 1001, 1):
    s = ''
    b = n
    while b != 0: 
        s = s + str(b%2)
        b = b//2
    s = s[::-1]
    s = s.replace('1', '0', 1) #    ()     1
    r = int(s, 2)
    d.append(n - r)
print(len(set(d)))
```

```python
def f(n):
    if n < 3:
        return n
    if n >= 3:
        return (n - 1) * f(n - 2)
print((f(2024)-f(2022))/f(2020))

# f(2024) - f(2022) = 2023*f(2022) - f(2022) = f(2022) * (2023 - 1) = f(2022) * 2022 = 2021 * f(2020) * 2022
```

```python
# 23 
# 7 -> 20, 14, !15
def f(x, y):
    if x > y and x == 15: #    15
        return 0
    if x == y:
        return 1
    else:
        return f(x+1, y) + f(x+3, y) + f(x*3, y)
    
print(f(7, 14) * f(14, 20)) # 14
```

```python
def f(x,y):
    if x < y: # and x == 36: #      (x == 36:)
        return 0
    if x == y:
        return 1
    else:
        return f(x - 2, y) + f(x // 2, y)
print(f(38, 16) *  f(16, 2))
```

```python
def f(x,y,s=''):
    if x > y or s[-2:] == '22':
        return 0
    if x == y:
        return 1
    else:
        return f(x+1,y,s+'1') + f(x+2,y,s+'2') + f(x*2,y,s+'3')
print(f(2, 22))
```

```python
#  ,   ,    .     . 
#    ,    . 
#                     . 
# ,    15 ,        16, 19  75 . 
#   ,   ,    . 
#     ,         68.

#   ,   ,     ,    68   . 
#       S ; 1  S  67.

#  ,     ,        . 
#    , ,        ,        . 
#             ,      , 
#         .

# 19) ,           .    S,    .

# 20)     S,       ,     :
#       ;
#         ,    .

# 21)    S,      :
#     ,           ;
#    ,       .


# >= 68   +1 + 4 *5

def v1(s):
    return (s+1>=68 or s+4>=68 or s*5>=68) and s<=67
def p1(s):
    return v1(s+1) and v1(s+4) and v1(s*5) and not(v1(s))
def v2(s):
    return (p1(s+1) or p1(s+4) or p1(s*5)) and not(v1(s))
def p2(s):                                             # 21
    return (v2(s+1) or v2(s+4) or v2(s*5)) and (v2(s+1) or v1(s+1)) and (v2(s+4) or v1(s+4)) and (v2(s*5) or v1(s*5)) and not(v1(s))


for s in range(1, 68):                                 # 20
    if v2(s) == 1:
        print(20, s)

for s in range(1, 68):                                 # 21
    if p2(s) == 1:
        print(21, s)
```
