```python
def f(a):
    print(a + 1)
# рекурсия - это вызов функции в самой себе, для нее нужно задать точку остановки, чтобы не делать её бесконечно 
def p(n):
    if n == 1:
        return 1
    else:
        return n + p(n-1)
```

```python
def f(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    if n>2:
        return f(n-2)*(n-1)
print(f(8))
```

```python
def f(n):
    if n == 0:
        return 0
    if n > 0 and n%2 == 0:
        return f(n//2)
    if n%2 == 1:
        return 1 + f(n-1)
c = 0
for n in range(1, 1001, 1):
    if f(n) == 3:
        c += 1
print(c)
```

```python
def f(n):
    if n == 1:
        return 1
    if n % 2 == 1 and n > 1:
        return n + f(n - 2)
    if n % 2 == 0:
        return n * f(n - 1)
print(f(40))
```
