```python
h = sorted('0123456789qwertyuiopasdfghjklzxcvbnm')
m = []
for x in h[0:9:1]:
    for y in h[0:9:1]:
        a = int('88'+x+'4'+y, 9) # range пишем с str
        b = int('7'+x+'44'+y, 11)
        if (a + b) % 61 == 0:
            c = a + b
            g = c // 61
            m.append(g)

print(min(m))  
```
