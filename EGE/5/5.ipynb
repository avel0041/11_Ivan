```python
# N ->   ->  ->  -> R

for n in range(128, 256, 1):
    s = ''
    b = n 
    while b != 0: #1
        s = s + str(b%2)
        b = b // 2
    s = s[::-1]
    s = s.replace('0','*') #2 *   ,   0  1  1  0
    s = s.replace('1','0')
    s = s.replace('*','1')
    #s.count('1') % 2 == 0    
    r = int(s, 2) #3
    if (n - r) == 185:
        print(n)
        break
```

```python
f = [] #    R\
for n in range(10000, -1, -1):
    s = ''
    b = n
    while b != 0: # b != 0
        s = s + str(b%4)
        b = b // 4
    s = s[::-1]
    v = list(map(int,list(s))) #list      sum      
    # list(s) -> ['1', '2', '3']; list(map(int,list(s))) -> [1, 2, 3]
    if sum(v) % 3 == 0:
        s = s.replace('0', '*')
        s = s.replace('2', '0')
        s = s.replace('*', '2')
        s = '32' + s 
        # 12345       5  0, 4  1, 3  2 ... (       (      ))(          )
    else:
        s = s + '33'
        s = s[:-4] + '01' + s[-2:]  #12{{3} 01 33}  <-- s[:-4] {3} == 123;    #123 01 33  <-- s[-2:] == 33
    r = int(s, 4)
    if r > 320:
        f.append(r)
    if r == 543: #   
        print(n)
        break

#print(min(f)) #543
```
