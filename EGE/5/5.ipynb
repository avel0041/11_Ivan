```python
# N ->   ->  ->  -> R

for n in range(128, 256, 1):
    s = ''
    b = n 
    while b != 0: #1
        s = s + str(b%2)
        b = b // 2
    s = s[::-1]
    s = s.replace('0','*') #2 *   ,   0  1  1  0
    s = s.replace('1','0')
    s = s.replace('*','1')
    #s.count('1') % 2 == 0    
    r = int(s, 2) #3
    if (n - r) == 185:
        print(n)
        break
```

```python
for n in range(1, 1000, 1):
    #   
    s = ''
    b = n 
    while b != 0:
        s = s + str(b%2)
        b = b // 2
    s = s[::-1]

    #A)
    j = 0                    
    for i in range(0, len(s), 1):
        j += int(s[i])
    if j % 2 == 0:
        s = s + '0'
    else:
        s = s + '1'

    #)
    j = 0
    for i in range(0, len(s), 1):
        j += int(s[i])
    if j % 2 == 0:
        s = s + '0'
    else:
        s = s + '1'
    #253
    r = int(s, 2)
    if r > 253:
        print(n)
        break

        
```
