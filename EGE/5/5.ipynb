```python
b = []
for n in range(1, 100000, 1):
    r = bin(n)[2:] # 1. bin в двоичную [2:] срез  
    #2.сумму цифр (двоичных), а не сумму всех чисел 1-100000 a)
    d = 0

    #'100010101' 1+0+0 el это цтфра из r
    for el in r:
        d = d + int(el) # преоброзовать el в число так как el строка 
    x = r + str(d%2) #r строка всегда кроме изначальной десятичной
    
    #б)
    d = 0 #обнуляем
    for el in x:
        d = d + int(el) # можно проще в двочной x.count('1') чтобы посчитать 1
    R = x + str(d%2)
    #преоброзовать r в десятичной для сравнения с 43
    R = int(R, 2)
    if R > 43:
        b.append(R) #дб в массив.append
 
print(min(b)) #min элемент в массиве
```

```python
for n in range(100, 1000000, 1):
    b =[]
    f = str(n)
    while len(f) != 2:
        b.append(int(f[0:3])) #int чтобы сделать массив не из строк 
        f = f[1:]
    s = min(b)
    d = max(b)
    o = d - s 
    if o == 623:
        print(n)
        break
        

```
