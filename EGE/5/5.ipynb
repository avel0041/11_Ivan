```python
b = []
for n in range(1, 100000, 1):
    r = bin(n)[2:] # 1. bin в двоичную [2:] срез  
    #2.сумму цифр (двоичных), а не сумму всех чисел 1-100000 a)
    d = 0

    #'100010101' 1+0+0 el это цтфра из r
    for el in r:
        d = d + int(el) # преоброзовать el в число так как el строка 
    x = r + str(d%2) #r строка всегда кроме изначальной десятичной
    
    #б)
    d = 0 #обнуляем
    for el in x:
        d = d + int(el) # можно проще в двочной x.count('1') чтобы посчитать 1
    R = x + str(d%2)
    R = int(R, 2) #преоброзовать r в десятичной для сравнения с 43
    if R > 43:
        b.append(R) #дб в массив.append
 
print(min(b)) #min элемент в массиве
```

```python
# N -> двоичная запись -> инверсия -> десятичную -> R

for n in range(128, 256, 1):
    s = ''
    b = n 
    while b != 0: #1
        s = s + str(b%2)
        b = b // 2
    s = s[::-1]
    s = s.replace('0','*') #2 * звездочка как буфер, чтобы поменять 0 на 1 а 1 на 0
    s = s.replace('1','0')
    s = s.replace('*','1')
    #s.count('1') % 2 == 0 если количество едениц четное
    r = int(s, 2) #3
    if (n - r) == 185:
        print(n)
        break
```

```python
f = [] # чтобы найти наименьший R\результат
for n in range(10000, -1, -1):
    s = ''
    b = n
    while b != 0: #пока b != 0
        s = s + str(b%4)
        b = b // 4
    s = s[::-1]
    v = list(map(int,list(s))) #list заносит в масив для функции sum которая работает только с масивом 
    # list(s) -> ['1', '2', '3']; list(map(int,list(s))) -> [1, 2, 3]
    if sum(v) % 3 == 0:
        s = s.replace('0', '*')
        s = s.replace('2', '0')
        s = s.replace('*', '2')
        s = '32' + s 
        # 12345 разряды идут с конца то есть 5 это 0, 4 это 1, 3 это 2 ... (полсе второй и третий это как индекс (если не пишут сторону авто справа ))(справа второй это не как индекс а просто второе число справа)
    else:
        s = s + '33'
        s = s[:-4] + '01' + s[-2:]  #12{{3} 01 33}  <-- s[:-4] {3} == 123; не включительно  #123 01 33  <-- s[-2:] == 33
    r = int(s, 4)
    if r > 320:
        f.append(r)
    if r == 543: # взяли из принта
        print(n)
        break

#print(min(f)) #543
```
