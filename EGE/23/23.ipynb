```python
# №23 
# 7 -> 20, 14, !15
def f(x, y):
    if x > y or x == 15: #число которое не содержит 15
        return 0
    if x == y:
        return 1
    else:
        return f(x+1, y) + f(x+3, y) + f(x*3, y)
    
print(f(7, 14) * f(14, 20)) #содержит 14
```

```python
def f(x,y):
    if x < y: # or x == 36: #не меняеться в зависимости от вычитания (x == 36:)
        return 0
    if x == y:
        return 1
    else:
        return f(x - 2, y) + f(x // 2, y)
print(f(38, 16) *  f(16, 2))
```

```python
def f(x,y,s):
    if x > y or s[-2:] == '22':
        return 0
    if x == y:
        return 1
    else:
        return f(x+1,y,s+'1') + f(x+2,y,s+'2') + f(x*2,y,s+'3')
print(f(2, 22, ''))
```
