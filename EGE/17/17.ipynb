```python
f = open('17.txt')
a = f.readlines()
d = list(map(int, a))
m = 0
for y in range(0, len(d), 1): #3)
    if d[y] > m and str(d[y])[-2::1] == '19':
        m = d[y]
troika = 0   
v = []
for i in range(0, len(d)-2, 1):
    c1 = d[i]
    c2 = d[i+1]
    c3 = d[i+2]

    #   1
    c = 0
    #3  (1.      ( ), 2.   ( ), 3.     )
    if c1 < 0:                     # 1() 
        if len(str(c1)) - 1 == 4:
            c+=1
    else:
        if len(str(c1)) == 4:
            c += 1
    if len(str(abs(c2))) == 4: # 2() abs( int)     
        c += 1
    if len(str(c3)) == 4:       # 3(b)
        c += 1
    if c == 2:
    
        #   2
        
        if c1 % 3 == 0 or c2 % 3 == 0 or c3 % 3 == 0:
            c123 = c1 + c2 + c3
            if c123 > m:
                troika += 1
                v.append(c123)

print(troika, max(v))
        
```
