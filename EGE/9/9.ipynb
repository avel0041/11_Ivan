```python
f = open('9(1).txt')
a = f.readlines()
c = 0
for el in a:
    b = el.split() #   split == \t, \n (          split: '1 2 3 4 5' -> ['1', '2', '3', ...])
    b = list(map(int, b)) #   int    b (    )
    s = sum(b)
    d = s - min(b) - max(b)
    d = d / 3
    if d >= 8:
        c += 1
print(c)
```
