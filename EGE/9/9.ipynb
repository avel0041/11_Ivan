```python
f = open('9(1).txt')
a = f.readlines()
c = 0
for el in a:
    b = el.split() #   split == \t, \n (          split: '1 2 3 4 5' -> ['1', '2', '3', ...])
    b = list(map(int, b)) #   int    b (    )
    s = sum(b)
    d = s - min(b) - max(b)
    d = d / 3
    if d >= 8:
        c += 1
print(c)
```

```python
f = open('9(2).txt')
a = f.readlines()
p = 0
for i in range(0, len(a), 1):
    # '1 2 3 4 5' -> a[i].split() -> ['1', '2', '3', '4', '5'] -> list(map(int, a[i].split)) -> [1, 2, 3, 4, 5]
    d = list(map(int, a[i].split()))
    if len(set(d)) == len(d):
        c = 0 #
        c1 = 0 #
        s = 0 # sum
        s1 = 0 # sum
        for m in d:
            if m % 2 == 1:
                c += 1
                s = s + m
            else:
                c1 += 1
                s1 = s1 + m
        if c > c1 and s < s1:
            p += 1
print(p)
    
```
