```python
f = open('9(1).txt')
a = f.readlines()
c = 0
for el in a:
    b = el.split() # автомотически определяет split == \t, \n (также можно настроить        split: '1 2 3 4 5' -> ['1', '2', '3', ...])
    b = list(map(int, b)) # применяем функцию int к всему массиву b (строки не складываются как числа)
    s = sum(b)
    d = s - min(b) - max(b)
    d = d / 3
    if d >= 8:
        c += 1
print(c)
```

```python
f = open('9(2).txt')
a = f.readlines()
p = 0
for i in range(0, len(a), 1):
    # '1 2 3 4 5' -> a[i].split() -> ['1', '2', '3', '4', '5'] -> list(map(int, a[i].split)) -> [1, 2, 3, 4, 5]
    d = list(map(int, a[i].split()))
    if len(set(d)) == len(d):
        c = 0 #нечет
        c1 = 0 #чет
        s = 0 #нечет sum
        s1 = 0 #чет sum
        for m in d:
            if m % 2 == 1:
                c += 1
                s = s + m
            else:
                c1 += 1
                s1 = s1 + m
        if c > c1 and s < s1:
            p += 1
print(p)
    
```

```python
f = open('9.txt')
a = f.readlines() 
c = 0
for i in range(0, len(a), 1):
    # d = list(map(int, a[i].split()))
    d = a[i].split() # -> ['37','83','24','19','37','41']  до a[0] = '37	83	24	19	37	41'
    v = []
    for el in d:
        v.append(int(el)) # list(map(int, a))  a = '12345' [1, 2, 3, 4, 5]; а list(map(int, (list(a)))) a = '12345' [1, 2, 3, 4, 5]
    p = [] # массив повторяющихся чисел
    n = [] # массив неповторяющихся чисел
    # v = [1, 1, 4, 5, 6, 6]
    for el in v:                    #  1 72 2 72
        if v.count(el) > 1:
            p.append(el)
        else:
            n.append(el)
    # p = [72, 72, 6, 6, 6] (если условие значение числа учитывается только 1 раз а не (столько раз, сколько оно встречается в строке)) -> [72, 6]
    if len(p) > 0 and len(n) > 0:
        if (sum(p) / len(p)) < (sum(n) / len(n)):
            c += 1
    # Если условие значение числа учитывается только 1 раз:
    # p = set(p)
    # if len(p) > 0 and len(n) > 0:
    #     if (sum(p) / len(p)) < (sum(n) / len(n)):
    #         c += 1
print(c)
```
