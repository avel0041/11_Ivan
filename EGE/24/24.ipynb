```python
a = open('24_8.txt').readline()
#a = 'xzctyvvbasdfghj'
f = []
c = 0
b = []
for i in range(0, len(a), 1):
    if a[i] in f:
        f = []
        b.append(c)
        c = 1
    else:
        f.append(a[i])
        c += 1
b.append(c)        #мы записываем последний счетчик (с) в массив b
print(max(b))
        
```

```python
a = open('24_2_1.txt').readline()
f = ''
c = 0 
b = []
for i in range(0, len(a)-3, 1): #* до len(a) - 3, потому что последний индекс i+3
    f = a[i:i+4:1]       #* i+4 не вкл это 3
    if f == 'XZZY':
        b.append(c)
        c = 0
    else:
        c += 1
print(max(b)+3)
```
