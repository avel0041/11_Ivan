```python
f = open('9.txt')
a = f.readlines() 
c = 0
for i in range(0, len(a), 1):
    # d = list(map(int, a[i].split()))
    d = a[i].split() # -> ['37','83','24','19','37','41']  до a[0] = '37	83	24	19	37	41'
    v = []
    for el in d:
        v.append(int(el)) # list(map(int, a))  a = '12345' [1, 2, 3, 4, 5]; а list(map(int, (list(a)))) a = '12345' [1, 2, 3, 4, 5]
    p = [] # массив повторяющихся чисел
    n = [] # массив неповторяющихся чисел
    # v = [1, 1, 4, 5, 6, 6]
    for el in v:                    #  1 72 2 72
        if v.count(el) > 1:
            p.append(el)
        else:
            n.append(el)
    # p = [72, 72, 6, 6, 6] (если условие значение числа учитывается только 1 раз а не (столько раз, сколько оно встречается в строке)) -> [72, 6]
    if len(p) > 0 and len(n) > 0:
        if (sum(p) / len(p)) < (sum(n) / len(n)):
            c += 1
    # Если условие значение числа учитывается только 1 раз:
    # p = set(p)
    # if len(p) > 0 and len(n) > 0:
    #     if (sum(p) / len(p)) < (sum(n) / len(n)):
    #         c += 1
print(c)
```

```python
for n in range(1, 1000, 1):
    # строим двоичную запись
    s = ''
    b = n 
    while b != 0:
        s = s + str(b%2)
        b = b // 2
    s = s[::-1]

    #A)
    j = 0                    
    for i in range(0, len(s), 1):
        j += int(s[i])
    if j % 2 == 0:
        s = s + '0'
    else:
        s = s + '1'

    #Б)
    j = 0
    for i in range(0, len(s), 1):
        j += int(s[i])
    if j % 2 == 0:
        s = s + '0'
    else:
        s = s + '1'
    #253
    r = int(s, 2)
    if r > 253:
        print(n)
        break

        
```
