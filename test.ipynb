```python
alp = sorted('1234567890qwertyuiopasdfghjklzxcvbnm')
for x in alp[:13]:
    G = int('4C'+x+'4', 15)
    G1 = int(x+'62A', 13)
    F = G + G1
    c = F % 121
    if c == 0:
        a = F / 121
        print(a)
        break
```
