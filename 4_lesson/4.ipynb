### Работа со строками

```python
s = 'Тестовая строка'
# + * not in len replace count ''.join(list) split

m = ['А', 'B', 'C']
m[0] = 13443123
print(m)


s = 'AAAABC'
s1 = 'D'
print(s[0])
print(s + s1)
print(s * 5)

if 'AB' in s: # если ab есть в строке то принт работает 
    print('АB есть в строке')

if 'D' not in s: # если d нету в строке то принт работает 
    print('B не находится в строчке s')

print(len(s)) # подсчет длины строки

print(s.count('A')) # посчитать сколько раз A в строке (s = строчка).count


l = list(s) # строка переделывается в массив
print(l)
l[-1] = 'Q'
print(l)

s2 = ''.join(l) # массив в строку
print(s2)

a = '1 25 3 4 59 6213124 7'
l1 = a.split()
print(l1)

s3 = 'AABBCC'
s3 = s3.replace('A', 'B', 1) # .replace(что заменить, на что заменить, сколько символов заменить)
print(s3)
```

```python
s = '8' * 68

while '222' in s or '888' in s:
    if '222' in s:
        s = s.replace('222', '8', 1) # .replace(что заменить, на что заменить, сколько символов заменить)
    else:
        s = s.replace('888', '2', 1)
print(s)

```

```python
s = '5' * 54 + '7'
while '722' in s or '557' in s:
    if '722' in s:
        s = s.replace('722', '57', 1)
    else:
        s = s.replace('557', '72', 1)
print(s)
```
