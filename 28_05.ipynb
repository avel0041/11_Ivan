```python
from itertools import product
a = list(product('КЗБС', repeat = 5))
c = 0
for i in range(0, len(a), 1):
    f = ''.join(a[i])
    if a[i].count('З') <= 2 and not('КБ' in f) and not('БК' in f):
        c += 1
print(c)
        
```

```python
f = open('9_1.txt')
a = f.readlines()
c = 0
for i in range(0, len(a), 1):
    d = list(map(int, a[i].split()))
    u1 = 0
    u2 = 0
    u3 = 0
    v1 = [] # числа, которые повторяются 3 раза
    v2 = [] # числа, которые встречаются 1 раз
    for el in d:
        if d.count(el) == 3:
            v1.append(el)
        elif d.count(el) == 1:
            v2.append(el)
    if len(v1) == 3 and len(v2) == 3:
        u1 = 1
    c1 = 0
    c2 = 0
    for el in d:
        if el % 2 == 0:
            c1 += 1
        else:
            c2 += 1
    if c1 > c2:
        u2 = 1
    b = sorted(d)
    if (b[4] + b[5]) > ((b[0] + b[1] + b[2] + b[3])*2): # (b[0] + b[1] + b[2] + b[3]) == sum(b[0:4])
        u3 = 1
    if u1 + u2 + u3 >= 2:
        c += 1
print(c)
    
```

```python
for n in range(4, 10000, 1):
    s = '0' + '5'*n
    while ('55' in s) or ('150' in s) or ('555' in s):
        if '55' in s:
            s = s.replace('55', '615', 1)
        if '150' in s:
            s = s.replace('150', '5950', 1)
        if '555' in s:
            s = s.replace('555', '50', 1)
    c = 1
    for el in s:
        if int(el) % 2 == 1:
            c *= int(el)
    if c > 100_000:
        print(n)
        break
```

```python
# ДЕЛ(n,m) -> n % m == 0
for A in range(1, 1000, 1):# (1<---, 1000, 1) A натуральное число
    f = 0
    for x in range(0, 16000, 1):
        if ((not(x % 100 == 0) and (x % 4 == 0)) or (x % 400 == 0) or not(x % A == 0)) == 0: # <-- #(т. е. принимает значение 1 при любом неотрицательном целом значении переменной х)
            f = 1
    if f == 0:
        print(A)
        break
```
