```python
for n in range(1000, 2, -1): # N   9876543 21                               for n in range(2, 1000, 1)             
    s = ''
    b = n                  # +
    while b!=0: 
        s = s + str(b%3)
        b = b//3           # +
    s = s[::-1]    # 
    if n % 3 == 0:
        s = s + s[-2::1] #  -2   (  2 )
    else:
        c = n % 3   
        c = c * 3
        g = ''
        while c!=0: 
            g = g + str(c%3)
            c = c//3
        g = g[::-1]
        s = s + g
    r = int(s, 3)
    if r <= 150:  # (<=) 150
        print(n)
        break
    
```
