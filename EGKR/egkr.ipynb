```python
for n in range(1000, 2, -1): # N больше чем 9876543 21                               for n in range(2, 1000, 1)             
    s = ''
    b = n                  # +
    while b!=0: 
        s = s + str(b%3)
        b = b//3           # +
    s = s[::-1]    #переворот 
    if n % 3 == 0:
        s = s + s[-2::1] # от -2 до конца (остается только 2 последних)
    else:
        c = n % 3   
        c = c * 3
        g = ''
        while c!=0: 
            g = g + str(c%3)
            c = c//3
        g = g[::-1]
        s = s + g
    r = int(s, 3)
    if r <= 150:  #не превышающее(<=) 150
        print(n)
        break
    
```

```python
from itertools import product
a = list(product('АБДЕОП', repeat=6))
for i in range(len(a)-1, -1, -1): #от конечного до начального ИНДЕКСА (100, -1, -1) (не это(100, 0, -1))
    if (i+1) % 2 == 0: #(i+1) из за того что список в задании начинаеться с 1, а (i) индексы в питоне с 0
        if a[i][0] == 'О' and a[i].count('А')==1 and a[i].count('Б')==1 and a[i].count('Д')==1 and a[i].count('Е')==1 and a[i].count('П')==1: # a[i](текущее слово с индексом i) [0](первая буква в этом слове из 6)
            if len(a[i]) == len(set(a[i])): #или так
                print(i+1)
                print(a[i])
                break
```
